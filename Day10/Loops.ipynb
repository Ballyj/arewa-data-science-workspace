# Day 10 Exercise: Loops

```python
# no.1
sum = 0
for i in range(10):
    print(i)
    
count = 0    
while count < 10:
    print(count)
    count = count + 1
    
# no.2    
sum = 10  
for i in range(10):
    sum = sum - 1
    print(sum)
    
count = 10
while count > 0:
    print(count)
    count = count - 1
    
# no. 3
for i in range(7):
    print((i + 1) * "#")
 
# no. 4   

for i in range(8):
    for j in range(8):
        print('#', end='  ')
    print()
    
# no. 5
n = 5
for i in range(1,11):
    print(n, '*', i, '=', n*i)
    
#no. 6
it_companies = ['Python', 'Numpy', 'Pandas', 'Django', 'Flask']
for  item in it_companies:
    print(item)

# no.7
for number in range(0, 100, 2):
    print(number)
    
#no.8
for number in range(1, 100+1, 2):
     print(number)
```

```python
# no. 1

sum = 1
for x in range(1,100):
    sum = sum + x
print('The sum of all numbers is: ', sum)

# no.2
sum = 0
for y in range(0, 100, 2):
    sum = sum + y
print('The sum of Even numbers is: ', sum)

sum = 0
for y in range(1, 100+1, 2):
    sum = sum + y
print('The sum of odd numbers is: ', sum)
```
